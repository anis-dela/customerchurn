```python
#Import Modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn')
```

```python
# Import Library Evaluation Model
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
```

- Data 1 : Data dengan Label Encoding untuk semua kolom kategorik
- Data 2 : Data dengan One Hot Encoding untuk semua kolom kategorik
- Data 3 : Data dengan kategori yang memiliki unique value <= 3 menggunakan Label Encoding , selain itu menggunakan One Hot Encoding
- Data 4 : Data dengan kategori yang memiliki unique value = 2 menggunakan Label Encoding , selain itu menggunakan One Hot Encoding

```python
# Import all dataset
data_1 = pd.read_csv('data_model1.csv')
data_2 = pd.read_csv('data_model2.csv')
data_3 = pd.read_csv('data_model3.csv')
data_4 = pd.read_csv('data_model4.csv')
```

```python
# Import test split
from sklearn.model_selection import train_test_split
```

```python
# Separate target and others data

X_1 = data_1.drop('Churn', axis=1)
X_2 = data_2.drop('Churn', axis=1)
X_3 = data_3.drop('Churn', axis=1)
X_4 = data_4.drop('Churn', axis=1)


y_1 = data_1['Churn']
y_2 = data_2['Churn']
y_3 = data_3['Churn']
y_4 = data_4['Churn']
```

```python
xtrain1, xtest1, ytrain1, ytest1 = train_test_split(X_1, y_1, test_size=0.2, random_state=41)
```

```python
xtrain2, xtest2, ytrain2, ytest2 = train_test_split(X_2, y_2, test_size=0.2, random_state=41)
```

```python
xtrain3, xtest3, ytrain3, ytest3 = train_test_split(X_3, y_3, test_size=0.2, random_state=41)
```

```python
xtrain4, xtest4, ytrain4, ytest4 = train_test_split(X_4, y_4, test_size=0.2, random_state=41)
```

### Random Forest Classifier

```python
# Import random forest
from sklearn.ensemble import RandomForestClassifier

rfc1 = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', max_depth = None, random_state = 41)
rfc2 = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', max_depth = None, random_state = 41)
rfc3 = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', max_depth = None, random_state = 41)
rfc4 = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', max_depth = None, random_state = 41)
```

```python
rfc1.fit(xtrain1, ytrain1)
rfc2.fit(xtrain2, ytrain2)
rfc3.fit(xtrain3, ytrain3)
rfc4.fit(xtrain4, ytrain4)
```

```python
ypred_rfc1 = rfc1.predict(xtest1)
ypred_rfc2 = rfc2.predict(xtest2)
ypred_rfc3 = rfc3.predict(xtest3)
ypred_rfc4 = rfc4.predict(xtest4)
```

```python
plot_confusion_matrix(rfc1, xtest1, ytest1, cmap = plt.cm.Blues)
plt.grid(False)
plt.title('Confussion Matrix for Data 1')

plot_confusion_matrix(rfc2, xtest2, ytest2, cmap = plt.cm.Blues)
plt.grid(False)
plt.title('Confussion Matrix for Data 2')

plot_confusion_matrix(rfc3, xtest3, ytest3, cmap = plt.cm.Blues)
plt.grid(False)
plt.title('Confussion Matrix for Data 3')

plot_confusion_matrix(rfc4, xtest4, ytest4, cmap = plt.cm.Blues)
plt.grid(False)
plt.title('Confussion Matrix for Data 4')
```

```python
# Generate Train, Test, and Cross Validation Score
rfc1_train_score = round(rfc1.score(xtrain1, ytrain1)*100, 2)
rfc1_test_score = round(rfc1.score(xtest1, ytest1)*100, 2)
rfc1_crossval = cross_val_score(rfc1, xtest1, ytest1, cv = 10).mean()

rfc2_train_score = round(rfc2.score(xtrain2, ytrain2)*100, 2)
rfc2_test_score = round(rfc2.score(xtest2, ytest2)*100, 2)
rfc2_crossval = cross_val_score(rfc2, xtest2, ytest2, cv = 10).mean()

rfc3_train_score = round(rfc3.score(xtrain3, ytrain3)*100, 2)
rfc3_test_score = round(rfc3.score(xtest3, ytest3)*100, 2)
rfc3_crossval = cross_val_score(rfc3, xtest3, ytest3, cv = 10).mean()

rfc4_train_score = round(rfc4.score(xtrain4, ytrain4)*100, 2)
rfc4_test_score = round(rfc4.score(xtest4, ytest4)*100, 2)
rfc4_crossval = cross_val_score(rfc4, xtest4, ytest4, cv = 10).mean()
```

```python
print('Classification Report Data 1')
print(classification_report(ytest1, ypred_rfc1))
print(f'Train Score : {rfc1_train_score} %')
print(f'Test Score : {rfc1_test_score} %')
print(f'Random Forest Classifier 1 Cross Validation Score : {round(rfc1_crossval*100, 2)} %')
print('\n')

print('Classification Report Data 2')
print(classification_report(ytest2, ypred_rfc2))
print(f'Train Score : {rfc2_train_score} %')
print(f'Test Score : {rfc2_test_score} %')
print(f'Random Forest Classifier 2 Cross Validation Score : {round(rfc2_crossval*100, 2)} %')
print('\n')

print('Classification Report Data 3')
print(classification_report(ytest3, ypred_rfc3))
print(f'Train Score : {rfc3_train_score} %')
print(f'Test Score : {rfc3_test_score} %')
print(f'Random Forest Classifier 3 Cross Validation Score : {round(rfc3_crossval*100, 2)} %')
print('\n')

print('Classification Report Data 4')
print(classification_report(ytest4, ypred_rfc4))
print(f'Train Score : {rfc4_train_score} %')
print(f'Test Score : {rfc4_test_score} %')
print(f'Random Forest Classifier 4 Cross Validation Score : {round(rfc4_crossval*100, 2)} %')
```
