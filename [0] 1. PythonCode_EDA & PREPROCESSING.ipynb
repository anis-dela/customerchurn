**5. Bagaimana korelasi Antara Variabel Kategorik dengan Churn?**

```python
data_kat = data.copy()
data_kat.drop(['tenure', 'MonthlyCharges', 'TotalCharges', 'customerID'], axis = 1, inplace = True)
data_kat
```

```python
import math
import scipy.stats as ss
from collections import Counter

def conditional_entropy(x,y):
    # entropy of x given y
    y_counter = Counter(y)
    xy_counter = Counter(list(zip(x,y)))
    total_occurrences = sum(y_counter.values())
    entropy = 0
    for xy in xy_counter.keys():
        p_xy = xy_counter[xy] / total_occurrences
        p_y = y_counter[xy[1]] / total_occurrences
        entropy += p_xy * math.log(p_y/p_xy)
    return entropy

def theil_u(x,y):
    s_xy = conditional_entropy(x,y)
    x_counter = Counter(x)
    total_occurrences = sum(x_counter.values())
    p_x = list(map(lambda n: n/total_occurrences, x_counter.values()))
    s_x = ss.entropy(p_x)
    if s_x == 0:
        return 1
    else:
        return (s_x - s_xy) / s_x
```

```python
# Correlation Between Churn and Categorical Variables

theilu = pd.DataFrame(index=['Churn'],columns=data_kat.columns)
columns = data_kat.columns
for j in range(0,len(columns)):
    u = theil_u(data_kat['Churn'].tolist(),data_kat[columns[j]].tolist())
    theilu.loc[:,columns[j]] = u
theilu.fillna(value=np.nan,inplace=True)
plt.figure(figsize=(20,1))
sns.heatmap(theilu,annot=True,fmt='.2f', cmap = 'YlGnBu')
plt.show()
```

### Encoding

Kombinasi encoding :
1. Pakai Label Encoding untuk semua kolom kategorik
2. Pakai One Hot Encoding untuk semua kolom kategorik
3. Untuk kategorik dengan 2/3 unique value menggunakan label encoding, sisanya pakai one hot encoding
4. Untuk kategorik dengan nilai 2 unique value hanya saja yang menggunakan label encoding, sisanya pakai one hot encoding
