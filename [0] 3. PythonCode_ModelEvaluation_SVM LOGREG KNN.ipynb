```python
#Import Modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
plt.style.use('seaborn')
```

```python
# Import all dataset
data_1 = pd.read_csv(r'data_model1.csv')
data_2 = pd.read_csv(r'data_model2.csv')
data_3 = pd.read_csv(r'data_model3.csv')
data_4 = pd.read_csv(r'data_model4.csv')
#ditulis datanya apa2 1 2 3 4 nya itu apa
#misal data encoded
```

#### List model
    1. Support Vector Machine
    2. Logistic Regression
    3. K-Nearest Neighbor

```python
# Import test split
from sklearn.model_selection import train_test_split
```

```python
data_1
```

```python
# Separate target and others data

X_1 = data_1.drop('Churn', axis=1)
X_2 = data_2.drop('Churn', axis=1)
X_3 = data_3.drop('Churn', axis=1)
X_4 = data_4.drop('Churn', axis=1)

y_1 = data_1[['Churn']]
y_2 = data_2[['Churn']]
y_3 = data_3[['Churn']]
y_4 = data_4[['Churn']]
```

```python
y_1
```

```python
X_train1, X_test1, y_train1, y_test1 = train_test_split(X_1, y_1, test_size=0.2, random_state=41)
```

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(X_2, y_2, test_size=0.2, random_state=41)
```

```python
X_train3, X_test3, y_train3, y_test3 = train_test_split(X_3, y_3, test_size=0.2, random_state=41)
```

```python
X_train4, X_test4, y_train4, y_test4 = train_test_split(X_4, y_4, test_size=0.2, random_state=41)
```

### Support Vector Machine

```python
from sklearn import svm
```

```python
svm1 = svm.SVC(C=1.0, kernel='rbf', degree=3, gamma='scale', random_state = 41)
svm2 = svm.SVC(C=1.0, kernel='rbf', degree=3, gamma='scale', random_state = 41)
svm3 = svm.SVC(C=1.0, kernel='rbf', degree=3, gamma='scale', random_state = 41)
svm4 = svm.SVC(C=1.0, kernel='rbf', degree=3, gamma='scale', random_state = 41)
```

```python
# Modelling and Train Data
svm1.fit(X_train1, y_train1)
svm2.fit(X_train2, y_train2)
svm3.fit(X_train3, y_train3)
svm4.fit(X_train4, y_train4)
```

```python
# Predict Y
y_pred_svm1 = svm1.predict(X_test1)
y_pred_svm2 = svm2.predict(X_test2)
y_pred_svm3 = svm3.predict(X_test3)
y_pred_svm4 = svm4.predict(X_test4)
```

```python
# Import Library Evaluation Model
from sklearn.metrics import plot_confusion_matrix
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
```

```python
# Evaluation Model
plot_confusion_matrix(svm1, X_test1, y_test1, cmap = plt.cm.Blues)
plt.title('Confussion Matrix for Data 1')
plt.grid(False)

plot_confusion_matrix(svm2, X_test2, y_test2, cmap = plt.cm.Blues)
plt.title('Confussion Matrix for Data 2')
plt.grid(False)

plot_confusion_matrix(svm3, X_test3, y_test3, cmap = plt.cm.Blues)
plt.title('Confussion Matrix for Data 3')
plt.grid(False)

plot_confusion_matrix(svm4, X_test4, y_test4, cmap = plt.cm.Blues)
plt.title('Confussion Matrix for Data 4')
plt.grid(False)

#WARNANYA GANTI BIAR FONTNYA BISA DIBACA
#TN FP
#FN TN

#TN : DIPREDIKSI TIDAK CHURN DAN BENAR TIDAK CHURN
#FN : DIPREDIKSI TIDAK CHURN PADAHAL CHURN
#FP : DIPREDIKSI CHURN PADAHAL TIDAK CHURN
#TP : DIPREDIKSI CHURN DAN BENAR CHURN
```

```python
# Generate Train, Test, and Cross Validation Score
svm1_train_score = round(svm1.score(X_train1, y_train1)*100, 2)
svm1_test_score = round(svm1.score(X_test1, y_test1)*100, 2)
svm1_crossval = cross_val_score(svm1, X_test1, y_test1, cv = 10).mean()

svm2_train_score = round(svm2.score(X_train2, y_train2)*100, 2)
svm2_test_score = round(svm2.score(X_test2, y_test2)*100, 2)
svm2_crossval = cross_val_score(svm2, X_test2, y_test2, cv = 10).mean()

svm3_train_score = round(svm3.score(X_train3, y_train3)*100, 2)
svm3_test_score = round(svm3.score(X_test3, y_test3)*100, 2)
svm3_crossval = cross_val_score(svm3, X_test3, y_test3, cv = 10).mean()

svm4_train_score = round(svm4.score(X_train4, y_train4)*100, 2)
svm4_test_score = round(svm4.score(X_test4, y_test4)*100, 2)
svm4_crossval = cross_val_score(svm4, X_test4, y_test4, cv = 10).mean()
```

```python
print('Classification Report Data 1')
print(classification_report(y_test1, y_pred_svm1))
print(f'Train Score : {svm1_train_score}%')
print(f'Test Score : {svm1_test_score}%')
print(f'Support Vector Machine 1 Cross Validation Score : {round(svm1_crossval*100, 2)}%')
print('\n')

print('Classification Report Data 2')
print(classification_report(y_test2, y_pred_svm2))
print(f'Train Score : {svm2_train_score}%')
print(f'Test Score : {svm2_test_score}%')
print(f'Support Vector Machine 1 Cross Validation Score : {round(svm2_crossval*100, 2)}%')
print('\n')

print('Classification Report Data 3')
print(classification_report(y_test3, y_pred_svm3))
print(f'Train Score : {svm3_train_score}%')
print(f'Test Score : {svm3_test_score}%')
print(f'Support Vector Machine 1 Cross Validation Score : {round(svm3_crossval*100, 2)}%')
print('\n')

print('Classification Report Data 4')
print(classification_report(y_test4, y_pred_svm4))
print(f'Train Score : {svm4_train_score}%')
print(f'Test Score : {svm4_test_score}%')
print(f'Support Vector Machine 1 Cross Validation Score : {round(svm4_crossval*100, 2)}%')
```
